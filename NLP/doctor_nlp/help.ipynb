```python
import json
```

```python
data = {}
with open("evidences/release_evidences.json", "r") as file:
    data = json.load(file)
```

```python
len([d for d, v in data.items() if len(v['possible-values']) > 0])
```

```python
relevant_data = {}
for key, val in data.items():
    rel_data_for_key = {}
    rel_data_for_key['question'] = val['question_en']
    rel_data_for_key['name'] = val['name']
    rel_data_for_key['value_meaning'] = {}
    for key_v_m, val_v_m in val['value_meaning'].items():
        rel_data_for_key['value_meaning'][key_v_m] = val_v_m['en']
    rel_data_for_key['possible-values'] = val['possible-values']
    relevant_data[key] = rel_data_for_key
```

```python
relevant_data
```

```python
with open("evidences/release_evidences_simplified.json", "w") as file:
    json.dump(relevant_data, file)
```

```python
with open("evidences/release_evidences_simplified.json", "r") as file:
    simplified_data = json.load(file)
```

```python
simplified_data_no_values = {key: value for key, value in simplified_data.items() if len(value['possible-values']) == 0}
simplified_data_no_value_meaning = {key: value for key, value in simplified_data.items() if len(value['possible-values']) > 0 and len(value['value_meaning']) == 0}
simplified_data_value_meaning = {key: value for key, value in simplified_data.items() if len(value['possible-values']) > 0 and len(value['value_meaning']) > 0}
```

```python
with open("evidences/release_evidences_simplified_no_values.json", "w") as file:
    json.dump(simplified_data_no_values, file)
```

```python
with open("evidences/release_evidences_simplified_no_value_meaning.json", "w") as file:
    json.dump(simplified_data_no_value_meaning, file)
```

```python
with open("evidences/release_evidences_simplified_value_meaning.json", "w") as file:
    json.dump(simplified_data_value_meaning, file)
```

```python
examples_no_values = {}
for key, val in simplified_data_no_values.items():
    question = val['question']
    example = ' '.join(question.split()[1:])
    example = example.replace('?', '')
    examples_no_values[key] = example
```

```python
examples_no_values
with open("evidences/release_evidences_simplified_no_values_examples.json", "w") as file:
    json.dump(examples_no_values, file)
```

```python
examples_no_value_meaning = {}

for key, val in simplified_data_no_value_meaning.items():
    question = val['question']
    scale = val['possible-values']
    examples_no_value_meaning[key] = {'example': question, 'scale': scale}
```

```python
examples_no_value_meaning
with open("evidences/release_evidences_simplified_no_value_meaning_examples.json", "w") as file:
    json.dump(examples_no_value_meaning, file)
```

```python
examples_value_meaning = {}
for key, val in simplified_data_value_meaning.items():
    question = val['question']
    print(question)
    example = input()
    codes_list = []
    q_c = {}
    for item in val['possible-values']:
        code = item + '(' + val['value_meaning'][item] + ')'
        codes_list.append(code)
    q_c['example'] = example
    q_c['codes'] = codes_list
    examples_value_meaning[key] = q_c
examples_value_meaning
```
