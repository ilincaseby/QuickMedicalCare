```python
import numpy as np
import matplotlib.pyplot as plt
import xgboost
```

```python
import pandas as pd
```

```python
df = pd.read_csv('symbipredict_2022.csv')
df
```

```python
df.sort_values(by='prognosis', inplace=True)
df
```

```python
symptoms = {}
index = 0
for item in df.keys().array[:-1]:
    symptoms[item] = index
    index += 1
print(symptoms)
diseases = {}
index = 0
for item in df['prognosis'].unique():
    diseases[item] = index
    index += 1
diseases
```

```python
import json

with open("symptoms_index_encoding.json", "w") as f:
    json.dump(symptoms, f, indent=4)

with open("diseases_index_encoding.json", "w") as f:
    json.dump(diseases, f, indent=4)
```

```python
known_diseases = set(df.prognosis.unique())
min_data_per_disease = 0
for disease in known_diseases:
    min_data_per_disease = max(min_data_per_disease, df.prognosis.value_counts().get(disease, 0))
test_length_per_disease = int(min_data_per_disease * .25)
```

```python
X_Train_list = []
Y_Train_list = []
X_Test_list = []
Y_Test_list = []
for disease in diseases.keys():
    y_t = np.zeros(len(known_diseases))
    y_t[diseases[disease]] = 1
    data = df.loc[df.prognosis == disease].sample(frac=1, random_state=1406).reset_index(drop=True)
    on_test = 0
    for index, row in data.iterrows():
        x_t = np.array(row[:-1])
        if on_test < test_length_per_disease:
            X_Test_list.append(x_t)
            Y_Test_list.append(y_t)
            on_test += 1
        else:
            X_Train_list.append(x_t)
            Y_Train_list.append(y_t)

print(len(X_Train_list), len(Y_Train_list), len(X_Test_list), len(Y_Test_list))
```

```python
X_Train = np.array(X_Train_list)
Y_Train = np.array(Y_Train_list)
X_Test = np.array(X_Test_list)
Y_Test = np.array(Y_Test_list)
```

```python
from sklearn.multioutput import MultiOutputClassifier
from sklearn.metrics import accuracy_score
```

```python
xgb_model = xgboost.XGBClassifier(objective='binary:logistic', eval_metric='logloss', use_label_encoder=False)
model = MultiOutputClassifier(xgb_model)
model.fit(X_Train, Y_Train)
Y_Pred = model.predict(X_Test)
accuracy_score(Y_Test, Y_Pred)
```

```python
from sklearn.svm import SVC
model_svc = SVC(kernel='linear', random_state=42)
multioutput_model = MultiOutputClassifier(model_svc)

multioutput_model.fit(X_Train, Y_Train)

Y_pred_svc = multioutput_model.predict(X_Test)
accuracy_score(Y_Test, Y_pred_svc)
```

```python
from sklearn.neural_network import MLPClassifier
model_mlp = MLPClassifier(hidden_layer_sizes=(100,), max_iter=1000, random_state=42)
multi_model_mlp = MultiOutputClassifier(model_mlp)

multi_model_mlp.fit(X_Train, Y_Train)
Y_pred_mlp = multi_model_mlp.predict(X_Test)
accuracy_score(Y_Test, Y_pred_mlp)
```
